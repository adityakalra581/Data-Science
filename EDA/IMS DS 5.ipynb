### BETTER APPROACH

using np.where()

more efficient even for alterations for multiple columns:

```python
import numpy as np
```

```python

# SYNTAX:
# dataset_name['new_column_name'] = np.where(dataset_name['c2']>,<,== any value , condition a, condition b)
# c2 = existing column in dataset on which condition is applied.
# condition a is executed, if the condition on c2 is true
# else: condition b is executed.
# see example below:


data['smart']= np.where(data['ActiveSinceDays']>500, 'IMS','CL')
data.head()
```
